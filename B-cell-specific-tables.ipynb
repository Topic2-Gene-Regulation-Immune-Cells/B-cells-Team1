## Import data

```python
import pandas as pd
ATAC = pd.read_csv("data/converted_ATAC_seq_info.csv", sep=";")
RNA = pd.read_csv("data/converted_ATAC_seq_info.csv", sep=";")
ATAC_QC = pd.read_csv("data/ATAC_QCmatric.CSV", sep=";")
ATAC_QC_RS = pd.read_csv("data/ATAC_QCmatric_ReadStatistics.CSV", sep=";")
```

Extraction of B cells and upstream hematopoietic stem and progenitor cells from the datasets

## ATAC-Seq data

```python
ATAC_Bcells = ATAC.iloc[:, :31]
ATAC_Bcells.to_csv("data/ATAC_Bcells.csv", index=False)
ATAC_Bcells.head()
```

## RNA-Seq data

```python
RNA_Bcells = RNA.iloc[:, :23]
RNA_Bcells.to_csv("data/RNA_Bcells.csv", index=False)
RNA_Bcells.head()
```

## ATAC_QC data

```python
ATAC_QC_Bcells = ATAC_QC.iloc[:44, :]
ATAC_QC_Bcells.to_csv("data/ATAC_QC_Bcells.csv", index=False)
ATAC_QC_Bcells.head()
```

## ATAC_QC_RS data

```python
ATAC_QC_RS_Bcells = ATAC_QC_RS.iloc[:42, :]
ATAC_QC_RS_Bcells.to_csv("data/ATAC_QC_RS_Bcells.csv", index=False)
ATAC_QC_RS_Bcells.head()
```

# ATAC-Seq transponierte Daten

```python
ATAC_Peaks_Bcells = ATAC.iloc[:, 8:31]
ATAC_T_Bcells = ATAC_Peaks_Bcells.T
ATAC_T_Bcells = ATAC_T_Bcells.reset_index()
ATAC_T_Bcells.to_csv("data/ATAC_T_Bcells.csv", index=False)
ATAC_T_Bcells.head()
```

# RNA-Seq transponierte Daten

```python
RNA_Bcells = RNA.iloc[:, :23]
RNA_T_Bcells = RNA_Bcells.T
RNA_T_Bcells = RNA_T_Bcells.reset_index()
RNA_T_Bcells.to_csv("data/RNA_T_Bcells.csv", index=False)
RNA_T_Bcells.head()
```
